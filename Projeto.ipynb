```python
# 1 -Top 10 países mais populosos?  

print(populacao_mundial.nlargest(10, 'Population_2026')[['Country', 'Population_2026']])
```

```python
# 2 - Quais países têm crescimento negativo?

print(populacao_mundial[populacao_mundial['Yearly_Change'] < 0][['Country', 'Yearly_Change']].sort_values('Yearly_Change'))
```

```python
# 3 - Qual a média de fertilidade por faixa de urbanização?

populacao_mundial['faixa_urbana'] = pd.cut(populacao_mundial['Urban_Population_pct'], bins=[0, 30, 50, 70, 100])
print(populacao_mundial.groupby('faixa_urbana')['Fertility_Rate'].mean())
```

```python
# 4 - Top 10 países mais densos?

print(populacao_mundial.nlargest(10, 'Density_per_km2')[['Country', 'Density_per_km2']])
```

```python
# 5 - Quais países mais recebem e emitem migrantes?

print(populacao_mundial.nlargest(5, 'Migrants_Net')[['Country', 'Migrants_Net']])
print(populacao_mundial.nsmallest(5, 'Migrants_Net')[['Country', 'Migrants_Net']])
```
